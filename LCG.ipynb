```python
import matplotlib.pyplot as plt
N=3000 # No. of iterations

a_1 = 811 #set of Multipliers
a_2 = 819
a_3 = 827

m_1 = 2**64 #set of modulus in form of 2^r
m_2 = 2**32
m_3 = 2**16

y_0_1=7   # Seed values which has to be odd for M=2^r and C=0
y_0_2=9
y_0_3=11


results = [] #Listing up the random numbers

for i in range(N):  # defining 3 lcgs
    y_0_1 = (a_1*y_0_1)%m_1
    y_0_2 = (a_2*y_0_2)%m_2
    y_0_3 = (a_3*y_0_3)%m_3
    
    x=(y_0_1 - y_0_2 + y_0_3)%(m_1-1) #combining 3 lcgs

    if x>0: # ondition for generation of random numbers in (0,1)
        R=x/m_1
    elif x==0:
        R=(m_1-1)/m_1    

    
    results.append(R)



#plotting
plt.plot(results," .b") 
plt.title("Random Number Generation combining 3 LCGs ")


plt.xlabel("Value of iterations upto N=3000")
plt.ylabel("Generated Random Numbers $(R_i)$")
plt.show()

```
