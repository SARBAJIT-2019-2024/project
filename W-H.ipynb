```python
#Wichman-Hill Generator

import matplotlib.pyplot as plt
N=3000 # No. of iterations

a_1 = 171 #set of Multipliers
a_2 = 172
a_3 = 170

m_1 = 30269 #set of modulus in form of 2^r
m_2 = 30307
m_3 = 30323

y_0_1=7   # Seed values which has to be odd for M=2^r and C=0
y_0_2=9
y_0_3=11


results = [] #Listing up the random numbers

for i in range(N):  # defining 3 lcgs
    y_0_1 = (a_1*y_0_1)%m_1
    y_0_2 = (a_2*y_0_2)%m_2
    y_0_3 = (a_3*y_0_3)%m_3
    
    u=(y_0_1/m_1 + y_0_2/m_2 + y_0_3/m_3) #combining 3 lcgs

    

    
    results.append(u)



#plotting
plt.plot(results," .b") 
plt.title("Wichman-Hill Generator ")


plt.xlabel("Value of iterations upto N=3000")
plt.ylabel("Generated Random Numbers $(R_i)$ between (0,3)")
plt.show()
```
