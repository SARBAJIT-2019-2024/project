```python
from pylab import plot,show
N=3600
a_1 = 800003
a_2 = 800011
a_3 = 800019

m_1 = 2**10
m_2 = 2**16
m_3 = 2**20

y_0_1=2**8+1
y_0_2=2**5+1
y_0_3=2**4+1


results = []

for i in range(N):
    y_0_1 = (a_1*y_0_1)%m_1
    y_0_2 = (a_2*y_0_2)%m_2
    y_0_3 = (a_3*y_0_3)%m_3
    
    x=(y_0_1 - y_0_2 + y_0_3)
    
    results . append(x)
plot (results," .r")

show()

```
